```python
def tran(text):
  url = 'https://translate.google.com/m?hl=zh-TW&sl=auto&tl=en&ie=UTF-8&prev=_m&q='
  wb_data = requests.get(url+urllib.parse.quote_plus(text))
  result = BeautifulSoup(wb_data.text,'lxml').select(".result-container")[0].get_text().replace("​","")
  return result
```

```python
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
import torch
model_id = "prompthero/openjourney"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")
```

```python
import edge_tts
import asyncio
from unsync import unsync
import os
import eyed3
from pydub import AudioSegment
from pydub.playback import play
import librosa
voice = 'zh-CN-YunyangNeural'
output = '4.mp3'
rate = '+5%'
volume = '+0%'
#def get_duration_mp3(file_path):
#    """
#    获取mp3音频文件时长
#    :param file_path:
#    :return:
#    """
#    try:
#      mp3Info = eyed3.load(file_path)
#      return mp3Info.info.time_secs
#    except AttributeError as e:
#      return 2
def change_pitch(percentage,path):
  sound = AudioSegment.from_file(path, format="mp3")
  print(sound.frame_rate)
  # shift the pitch down by half an octave (speed will decrease proportionally)
  octaves = percentage
  new_sample_rate = int(sound.frame_rate * (2.0 ** octaves))
  lowpitch_sound = sound._spawn(sound.raw_data, overrides={'frame_rate': new_sample_rate})
  #Play pitch changed sound
  lowpitch_sound.export(path, format="mp3")
@unsync
async def my_function(TEXT):
    tts = edge_tts.Communicate(text = TEXT,voice = voice,rate = rate,volume=volume)
    T = TEXT.replace("/","")
    await tts.save("./a/"+T+".mp3")
    while not os.path.exists("./a/"+T+".mp3"):
      pass
    change_pitch(0.15,"./a/"+T+".mp3")
    return librosa.get_duration(path="./a/"+T+".mp3")
my_function("其实导致这个问题的主要原因是因为文件下的文件有问题，可能是其他软件安装过程或是其他原因导致的，这里删除掉然后重建即可。").result()
```
